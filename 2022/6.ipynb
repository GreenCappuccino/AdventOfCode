```python
data = open("data.txt", "r").read().strip()
```

```python
from collections import deque
```

```python
window = deque()
i = 0
for datum in data:
    window.appendleft(data[i])
    if len(window) > 4:
        window.pop()
    print(window)
    # Check if it is mark
    if len(window) == 4 and len(set(window)) == len(window):
        break
    i+=1
```

```python
i+1
```
