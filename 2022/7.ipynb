```python
data = open("data.txt", "r").read()
```

```python
lines = data.splitlines()
lines.reverse()
```

```python
from typing import Optional
from dataclasses import dataclass
@dataclass
class Node:
    size: int
    parent: any
    children: Optional[list]
    isDir: bool
```

```python
def propagate_size(node: Node):
    add = node.size
    held = node.parent
    while held.parent is not None:
        held.size += add
        held = held.parent
    held.size += add
```

```python
stack = []
root = None
while len(lines) > 0:
    print(len(lines))
    line = lines.pop()
    cmd = line[2:4]
    print("cmd",cmd)
    if cmd == "cd":
        if line[5:].strip() != "..":
            newDir = Node(0,None,[],True)
            if len(stack) > 0:
                newDir.parent = stack[-1]
                stack[-1].children.append(newDir)
            stack.append(newDir)
        else:
            stack.pop()
    elif cmd == "ls":
        while len(lines) > 0 and lines[-1][0] != "$":
            size_str, file = lines.pop().split(" ")
            print("size_str", size_str)
            print("file", file)
            if size_str != "dir":
                fileNode = Node(int(size_str), stack[-1], None, False)
                stack[-1].children.append(fileNode)
                propagate_size(fileNode)
    root = stack[0]
```
