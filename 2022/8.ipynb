```python
data = open("data.txt", "r").read()
```

```python
lines = data.splitlines()
```

```python
height = len(lines)
```

```python
width = len(lines[0].strip())
```

```python
all = ""
for line in lines:
    all += line.strip()
```

```python
# set of indexes of visible trees
visible = set()
```

```python
# rows (LtR)
for i in range(0, height):
    # grab edge
    tallest = all[i*width]
    visible.add(i*width)
    # need to iterate column-wise
    for j in range(1, width - 1):
        if all[i*width + j] > tallest:
            tallest = all[i*width + j]
            visible.add(i*width + j)
```

```python
# rows (RtL)
for i in range(0, height):
    # grab edge
    tallest = all[i*width+width-1]
    visible.add(i*width+width-1)
    # need to iterate column-wise
    for j in range(width - 2, 0, -1):
        if all[i*width + j] > tallest:
            tallest = all[i*width + j]
            visible.add(i*width + j)
```

```python
# columns (TtB)
for i in range(0, width):
    # grab edge
    tallest = all[i]
    visible.add(i)
    # need to iterate row-wise
    for j in range(1, width - 1):
        if all[j*width+i] > tallest:
            tallest = all[j*width + i]
            visible.add(j*width + i)
```

```python
# columns (BtT)
for i in range(0, width):
    # grab edge
    tallest = all[width*(height-1)+i]
    visible.add(width*(height-1)+i)
    # need to iterate row-wise
    for j in range(width - 2, 0, -1):
        if all[j*width+i] > tallest:
            tallest = all[j*width + i]
            visible.add(j*width + i)
```

## Part 2

```python
def scenic(row,col):
    print("----------")
    # rows (LtR)
    tree_height = int(all[row*width + col])
    ltr_num = -1
    # need to iterate column-wise
    for i in range(row*width + col, row*width + col + (width-col)):
        print("ltr", all[i])
        ltr_num += 1
        if int(all[i]) >= tree_height and i != row*width + col:
            break
    # rows (RtL)
    rtl_num = -1
    # need to iterate column-wise
    for i in range(row*width + col, row*width-1, -1):
        print("rtl", all[i])
        rtl_num += 1
        if int(all[i]) >= tree_height and i != row*width + col:
            break
    # columns (TtB)
    ttb_num = -1
    # need to iterate column-wise
    for i in range(row*width + col, width * height, width):
        print("ttb", all[i])
        ttb_num += 1
        if int(all[i]) >= tree_height and i != row*width + col:
            break
    # columns (BtT)
    btt_num = -1
    # need to iterate column-wise
    for i in range(row*width + col, -1, -width):
        print("btt", all[i])
        btt_num += 1
        if int(all[i]) >= tree_height and i != row*width + col:
            break
    if row*width + col == 17:
        print(ltr_num, rtl_num, ttb_num, btt_num)
    return ltr_num*rtl_num*ttb_num*btt_num
```

```python
scenic_scores = []
for i in range(height):
    for j in range(width):
        scenic_scores.append(scenic(i, j))
```
